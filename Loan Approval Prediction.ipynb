<h3 align="left"><font color="navy"> <center> Datamizni to'ldirish uchun biz imblearn kutubxonasidagi SMOTENC metdodidan foydalanamiz. Bu metod SMOTE(Synthetic Minority Oversampling Technique) - oilasiga mansub bo'lib kam classlarni to'ldirish uchun xizmat qiladi. U bir xil sinfning tasodifiy datalari orasidagi chiziqli interpolyatsiyadan foydalangan holda ishlaydi. Ya'nada soddaroq aytadigan bo'lsak u knn(k ta eng yaqin qo'shnilar) orqali ishlaydi. Ya'ni uning qo'shnilari orqali yangi datalar hosil qilinadi.</font></h3> 


</div>

```python
from imblearn.over_sampling import SMOTENC
```

```python
# 1-bo'lib. Datani X va y ga ajratib olamiz
X = data.drop('loan_status', axis=1)  
y = data['loan_status'] 
```

```python
# Object ustunlarni indexini olamiz
categorical_features_indices = [X.columns.get_loc(col) for col in categorical_columns]
```

```python
# SMOTENC-dan obyektini yaratib olamiz
# biz object ustunlari Numeric qilgan bo'lsak ham SMOTENC-ga bular haqida ko'ratishimiz kerak bo'ladi.
smote_nc = SMOTENC(categorical_features=categorical_features_indices, random_state=42, k_neighbors=5)
```

```python
# SMOTENC ishlatamiz
X_resampled, y_resampled = smote_nc.fit_resample(X, y)
```

```python
# Kerali ustunlarni datamiz oldin qanday formatda yaxlitlangan bo'lsa xuddi shunday yaxlitlashga o'tkazamiz
X_resampled['person_emp_length'] = X_resampled['person_emp_length'].round(0)  # nuqtadan keyin 0 tagacha
X_resampled['loan_int_rate'] = X_resampled['loan_int_rate'].round(2)         # nuqtadan keyin 2 tagacha
X_resampled['loan_percent_income'] = X_resampled['loan_percent_income'].round(2)  # nuqtadan keyin 2 tagacha
```

```python
X_resampled
```

```python
# Datamizni yaxshilab o'rganish uchun oldingi holatiga qaytarib olamiz
for col in categorical_columns:
    X_resampled[col] = label_encoders[col].inverse_transform(X_resampled[col])
```

```python
X_resampled
```

```python
# umumiy datasetga yangi yaratgan datasetimizni qo'shamiz
data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['loan_status'])], axis=1)
```
